```python
min_out = []
max_out = []
for i in range(len(diabetes.columns)):
    Q1 = diabetes.iloc[:,i].quantile(0.25)
    Q3 = diabetes.iloc[:,i].quantile(0.75)
    IQR = Q3 - Q1
    min_outlier = Q1 - 1.5*IQR
    max_outlier = Q3 + 1.5*IQR
    min_out.append(min_outlier)
    max_out.append(max_outlier)





print(min_out)
print(max_out)
```

```python
diabetes_no_outliers = diabetes
for x in range(len(diabetes.columns)):
    diabetes_no_outliers = diabetes_no_outliers[(diabetes_no_outliers.iloc[:,x]<=max_out[x]) & (diabetes_no_outliers.iloc[:,x]>=min_out[x])]

diabetes_no_outliers
```

```python
plt.boxplot(diabetes_no_outliers['Age'])
plt.show()
```

# feature Engineering

# Scaling data

```python
X = diabetes_no_outliers.iloc[:,:-1].values
X
```

```python
y = diabetes_no_outliers.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scal = MinMaxScaler()
```

```python
X_scaled = scal.fit_transform(X)
X_scaled
```

# spliting data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Create Model

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
y_pred = log.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(100*accuracy_score(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```
